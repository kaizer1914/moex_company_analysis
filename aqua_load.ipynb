 Загрузка файла из https://russaquaculture.ru/shareholders-and-investors/reports-and-publications/financial-statements/

```python
import pandas as pd

url = 'https://russaquaculture.ru/upload/iblock/58a/3qxfuim51d31d4g54p8g16dvmi0b33eg.xlsx'

aqua_balance = pd.read_excel(url, sheet_name='Balance sheet')
aqua_balance.drop('Unnamed: 0', axis='columns', inplace=True)
aqua_balance.rename(columns={'Unnamed: 1': 'chapter'}, inplace=True)
aqua_balance = aqua_balance[aqua_balance['chapter'].notna()]
aqua_balance.reset_index(drop=True, inplace=True)
aqua_balance
```

Сохраняем датафрем в БД

```python
%load_ext sql
%sql sqlite:///:memory:
%sql --persist aqua_balance
```
